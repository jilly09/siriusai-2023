В папке с проектом должны быть файлы train_1.json, (...) train_20.json, а также файлы test_00.json и test_01.json

```python
import json
dialogs = {}
```

Получаем словарь, где ключ - 3 последних сообщения в диалоге до сообщения отправки фото, а значение - ссылка на это фото.

```python
for i in range(0, 21):
  with open(f"train_{i}.json", "r") as file:
    object1 = json.loads(file.read())
    for dialogue in object1:
      for i in range(len(dialogue["dialogue"])):
        if dialogue["dialogue"][i]["share_photo"]:
          string = dialogue["dialogue"][i-2]["message"] + dialogue["dialogue"][i-1]["message"] + dialogue["dialogue"][i]["message"]
          dialogs[string] = dialogue["photo_url"]

print(dialogs)
```

Для тестовых данных:

```python
for i in range(0, 2):
  with open(f"test_0{i}.json", "r") as file:
    object1 = json.loads(file.read())
    for dialogue in object1:
      for i in range(len(dialogue["dialogue"])):
        if dialogue["dialogue"][i]["share_photo"]:
          string = dialogue["dialogue"][i-2]["message"] + dialogue["dialogue"][i-1]["message"] + dialogue["dialogue"][i]["message"]
          dialogs[string] = dialogue["photo_url"]

print(dialogs)
```

Получаем произвольный диалог и последние три реплики (final_string).<br>
1 input = 1 сообщение

```python
context = []
while True:
  message = input("Enter a message: ")
  context.append(message)
  if "[Share the photo]" in message:
    break

final_string = context[-3]+context[-2]+context[-1].replace("[Share the photo]", "")
print(final_string)
```

Выбираем наиболее похожую реплику на данную нам (final_string) из множества dialogs при помощи библиотеки rapidfuzz.

```python
import rapidfuzz
compairing = rapidfuzz.process.extract(final_string, dialogs.keys(), limit=10)
print(compairing)
print('\n')

max_raiting = None
max_link = None
ap = 0
j = 0
x = 0
for i in compairing:
  print(f"Phrases: {i[0]}\nLink: {dialogs[i[0]]}")
  ask = int(input("Do you like this option (0/1)?"))
  if ask == 1 and (max_raiting == None or (max_raiting < 1+i[1]/100)):
    print("ok")
    max_raiting = 1+(i[1]/100)
    max_link = dialogs[i[0]]
  if ask == 1:
    x = x + 1
  j = j + 1
  ap = ap + (x / j) * ask

print('\n')
print(max_raiting, max_link)
print(ap / 10)
```
